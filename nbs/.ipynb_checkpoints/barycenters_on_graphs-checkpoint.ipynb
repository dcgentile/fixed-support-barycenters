```python
import numpy as np
import pandas as pd
import networkx as nx
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm
import torch
from numba import jit
```

```python
@jit
def potential_energy(V: np.ndarray, mu:np.ndarray) -> float:
    return np.dot(V, mu)

@jit
def entropy(mu: np.ndarray) -> float:
    lnmu = np.array([0 if np.isclose(x,0) else np.log(x) for x in mu])
    
    return (-mu.T @ lnmu)

@jit
def numerical_directional_derivative(f:callable, 
                                     mu: np.ndarray, 
                                     nu: np.ndarray, 
                                     h:float, 
                                    ) -> float:
    perturbation = mu + h*(nu - mu)
    delta = f(perturbation) - f(mu)
    if np.isnan(delta):
        print(perturbation, delta)
        return 0
    else:
        return delta / h

@jit
def numerical_frechet_derivative(f:callable, mu:np.ndarray, h:float) -> np.ndarray:
    N = mu.size
    delta_f = np.zeros((N,1))
    I = np.identity(N)
    for i in range(N):
        delta_f[i,0] = numerical_directional_derivative(f, mu, I[:,i], h)
    return delta_f - np.average(delta_f)

@jit
def G_gradient(A: np.ndarray, f: np.ndarray) -> np.ndarray:
    N, _ = A.shape
    diffs = np.zeros((N,N), dtype=np.float64)
    for i in range(N):
        for j in range(i, N):
            dij = f[i] - f[j]
            dji = f[j] - f[i]
            diffs[i, j] = dij.item()
            diffs[j, i] = dji.item()

    return A * diffs

@jit
def logarithmic_mean(r,s):
    if np.isclose(r,s):
        return r
    else:    
        return (r - s) / (np.log(r) - np.log(s))

@jit
def metric_tensor(rho: np.ndarray, g=logarithmic_mean) -> np.ndarray:
    N = rho.size
    metric = np.zeros((N,N))
    for i in range(N):
        for j in range(i+1, N):
            gij = g(rho[i], rho[j])
            metric[i,j] = gij
            metric[j,i] = gij
    return metric

def random_simplex(N: int, seed=0) -> np.ndarray:
    rng = np.random.default_rng(seed=seed)
    v = rng.random(N)
    return (v / np.linalg.norm(v,1))

@jit
def uniform_measure(N: int) -> np.ndarray:
    v = np.ones(N)
    return (v / np.linalg.norm(v, 1))

@jit
def rho_inner_product(A:np.ndarray, v: np.ndarray, u: np.ndarray, rho: np.ndarray, g=logarithmic_mean) -> float:
    metric = metric_tensor(rho, g)
    product = A * v * u * metric
    return (1/2) * np.sum(product)

@jit
def sq_rho_norm(A:np.ndarray, v:np.ndarray, rho:np.ndarray, g=logarithmic_mean) -> float:
    return rho_inner_product(A, v, v, rho, g)

@jit
def div_rho(A:np.ndarray, v: np.ndarray, rho: np.ndarray, g=logarithmic_mean) -> np.ndarray:
    N, _ = v.shape
    u = np.zeros((N,1))
    g = metric_tensor(rho, g)
    for i in range(N):
        vi = v[:,i]
        ai = A[:,i]
        gi = g[i,:]
        u[i,0] = np.sum(vi * ai * gi)
    return u


#found this projection code from GH user daien
#https://gist.github.com/daien/1272551/edd95a6154106f8e28209a1c7964623ef8397246
@jit
def euclidean_proj_simplex(v, s=1):
    """ Compute the Euclidean projection on a positive simplex
    Solves the optimisation problem (using the algorithm from [1]):
        min_w 0.5 * || w - v ||_2^2 , s.t. sum_i w_i = s, w_i >= 0 
    Parameters
    ----------
    v: (n,) numpy array,
       n-dimensional vector to project
    s: int, optional, default: 1,
       radius of the simplex
    Returns
    -------
    w: (n,) numpy array,
       Euclidean projection of v on the simplex
    Notes
    -----
    The complexity of this algorithm is in O(n log(n)) as it involves sorting v.
    Better alternatives exist for high-dimensional sparse vectors (cf. [1])
    However, this implementation still easily scales to millions of dimensions.
    References
    ----------
    [1] Efficient Projections onto the .1-Ball for Learning in High Dimensions
        John Duchi, Shai Shalev-Shwartz, Yoram Singer, and Tushar Chandra.
        International Conference on Machine Learning (ICML 2008)
        http://www.cs.berkeley.edu/~jduchi/projects/DuchiSiShCh08.pdf
    """
    #assert s > 0, "Radius s must be strictly positive (%d <= 0)" % s
    n, = v.shape  # will raise ValueError if v is not 1-D
    # check if we are already on the simplex
    if v.sum() == s and np.all(v >= 0):
        # best projection: itself!
        return v
    # get the array of cumulative sums of a sorted (decreasing) copy of v
    u = np.sort(v)[::-1]
    cssv = np.cumsum(u)
    # get the number of > 0 components of the optimal solution
    rho = np.nonzero(u * np.arange(1, n+1) > (cssv - s))[0][-1]
    # compute the Lagrange multiplier associated to the simplex constraint
    theta = float(cssv[rho] - s) / rho
    # compute the projection by thresholding v using theta
    w = np.clip((v - theta), a_min=0, a_max=1)
    return w


def test_ibp(seed=0, N=10, generator=nx.cycle_graph):
    failed_tests = 0
    rng = np.random.default_rng(seed=seed)
    G = generator(N)
    A = nx.to_numpy_array(G)
    phi = N * rng.random(N)
    psi = N * rng.random(N)
    rho = random_simplex(N)
    grad_phi = G_gradient(A, phi)
    grad_psi = G_gradient(A,psi)
    lhs = rho_inner_product(A, grad_phi, grad_psi, rho)
    rhs = phi.T @ div_rho(A, grad_psi, rho)
    assert not np.isclose(rhs, 0)
    assert not np.isclose(lhs, 0)
    if not np.isclose(lhs+rhs,0):
        failed_tests += 1
    return failed_tests

def distance_matrix(G: nx.graph) -> np.ndarray:
    distances = dict(nx.shortest_path_length(G))
    distances = np.array([[list(distances[key].values())] for key in distances.keys()]).squeeze()
    return distances
```

```python
N = 5
h = 1e-1
G = nx.cycle_graph(N)
iters = 2**10
V = np.arange(N)
distances = distance_matrix(G)
f = entropy
A = nx.to_numpy_array(G)

rho = random_simplex(N)
rho_steps = np.zeros((iters, N))
for i in tqdm(range(iters)):
    delta_f = numerical_frechet_derivative(f, rho, h)
    wass_grad_f = G_gradient(A, delta_f)
    divergence = div_rho(A, wass_grad_f, rho)
    step = rho - h * divergence.T.squeeze()
    rho = euclidean_proj_simplex(step)
    rho_steps[i, :] = rho
```

```python
fig, ax = plt.subplots() 
df = pd.DataFrame(rho_steps, columns=np.arange(N))
df.head()
for i in range(N):
    sns.lineplot(data=df, x=df.index, y=i)
ax.set_xlabel('Step')
ax.set_ylabel('Mass')
ax.set_title(f'Mass per node under the gradient flow of a entropy')
plt.show()
plt.clf()

fig, ax = plt.subplots(1,3)
fig.set_size_inches(15,5)
idx = (0, iters // 2, iters -1)
vmin, vmax, cmap = 0, np.max(rho_steps[0,:]), "magma"
for i in range(3):
    nx.draw(G, 
            ax=ax[0], 
            node_color=rho_steps[idx[0],:], 
            vmin=vmin, 
            vmax=vmax,
            cmap=cmap,
            seed=0
    )
```
